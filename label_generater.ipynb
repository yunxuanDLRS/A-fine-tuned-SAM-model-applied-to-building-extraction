```python
# 将分割图和原图合在一起
from PIL import Image
import cv2
import numpy as np
import torch
import matplotlib.pyplot as plt
import os


path_image = 'C:/work/IMG/BJ1225/val/image_seg_1024/'
path_s_image = 'C:/work/IMG/BJ1225/val/image_sin_1024/'
path_label = 'C:/work/IMG/BJ1225/val/label_seg_1024/'
path_s_label = 'C:/work/IMG/BJ1225/val/label_sin_1024/'

num = len(os.listdir(path_label))
print(num)
filelist = os.listdir(path_label)
    
class UnionFind:
    def __init__(self):
        # 初始化并查集，使用字典来存储每个节点的父节点
        self.parent = {}

    def find(self, x):
        # 查找节点所属的集合的根节点
        if x not in self.parent:
            # 如果节点不存在于字典中，说明它是一个独立的集合，将其父节点设为自身
            self.parent[x] = x
            return x
        elif self.parent[x] == x:
            # 如果节点的父节点是自身，说明已找到根节点
            return x
        else:
            # 递归调用find函数找到根节点，并进行路径压缩
            self.parent[x] = self.find(self.parent[x])
            return self.parent[x]

    def union(self, x, y):
        # 合并两个节点所属的集合
        root_x = self.find(x)
        root_y = self.find(y)
        if root_x != root_y:
            # 如果两个节点不属于同一集合，将一个集合的根节点设为另一个集合的根节点
            self.parent[root_x] = root_y

def split_connected_components(coordinates):
    # 初始化并查集
    uf = UnionFind()

    for coord in coordinates:
        x, y = coord
        # 将相邻的节点进行合并
        uf.union(coord, (x + 1, y))
        uf.union(coord, (x - 1, y))
        uf.union(coord, (x, y + 1))
        uf.union(coord, (x, y - 1))

    # 组织不同集合的节点
    components = {}
    for coord in coordinates:
        root = uf.find(coord)
        if root not in components:
            components[root] = [coord]
        else:
            components[root].append(coord)

    return list(components.values())


for item in filelist:
    # 打开图片
    image = Image.open(path_label + item)
    image = np.array(image)
    print(image.shape)
    prompt_point_indices = np.argwhere(image[:,:,0] != 0) 
    result_list = [tuple(row) for row in prompt_point_indices]
    result = split_connected_components(result_list)
    print(len(result))
    for i in range(len(result)):
        tempImg = np.zeros([1024,1024])
        result_path = path_s_label + item.split(".")[0] + '_{}.tif'.format(i)
        for position in result[i]:
            tempImg[position] = 255
        binary_image = cv2.cvtColor(tempImg.astype(np.uint8), cv2.COLOR_GRAY2BGR)
        cv2.imwrite(result_path, binary_image)
        
        img = Image.open(path_image + item)
        img.save(path_s_image + item.split(".")[0] + '_{}.tif'.format(i))
    
        print('图片{}的第{}个对象提取完成'.format(item,i))
```
